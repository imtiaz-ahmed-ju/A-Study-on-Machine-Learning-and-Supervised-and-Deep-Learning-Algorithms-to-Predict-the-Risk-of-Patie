#Feature Importance

```python
x = df1.iloc[:,:-1]
```

```python
x
```

```python
y = df1['TenYearCHD']
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
fit_features = SelectKBest(score_func=f_classif)
```

```python
fit_features.fit(x,y)
```

```python
pd.DataFrame(fit_features.scores_)
```

#Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.30, random_state=47)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lreg = LinearRegression()
```

```python
Lreg.fit(xtrain,ytrain)
```

```python
Lreg.predict(xtest)
```

```python
Lreg.score(xtest,ytest)
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
Loreg = LogisticRegression()
```

```python
Loreg.fit(xtrain,ytrain)
```

```python
Loreg.predict(xtest)
```

```python
Loreg.score(xtest,ytest)
```

#ROC Curve Logistic Regression

```python
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn import datasets
```

```python
Loreg_y_score = Loreg.predict_proba(xtest)[:,1]
```

```python
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(ytest, Loreg_y_score)
```

```python
print('roc_auc_score for Logistic Regression: ', roc_auc_score(ytest, Loreg_y_score))
```

```python
plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic - Logistic regression')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
